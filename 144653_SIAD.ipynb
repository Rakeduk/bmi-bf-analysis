<h2>Establishment of Problem

This amount of data allows Us to perform certain analyst and confirm/deny different theories. We will try to <i> break</i> popular body measurement which is used to control correct weight and is even used in professional dietician. This indicator is called BMI (<i>body mass index</i>).

---

To better understand how BMI work we should look <i>inside it</i>:

$$ BMI = \frac{weight}{height^2}  =  [\frac{KG}{M^2}]$$

```r
BMI_levels <-c("Below 18.5", "18.5 - 24.9", "25.0 - 29.9", "30.0 and Above" )
names (BMI_levels) <-c("Underweight", "Normal", "Overweight", "Obese")
BMI_levels
```

The problem with this <i>BMI</i> indicator is the fact that it doesn't include the level of your fitness. To present it better:

![Obraz](https://healthtravelguide.com/wp-content/uploads/2020/06/Same-BMIs-different-body-types.png)

<h2>How to measure the level of fitness?</h2>

The best way to measure fitness of different people would be physical efficiency test, however as we don't have this type of information, we will try to predict level of fitness with second most common body indicator ~ <b>percentage of body fat</b>
 


$$  Body fat_{man}  = \frac{1.634 *  waist - 0.1804 * weight - 98.42}{2.2 * weight} * 100 $$

$$  Body fat_{woman} = \frac{1.634 *  waist - 0.1804 * weight - 76.76}{2.2 * weight} * 100 $$

---

Our task is not easy, because in contrast to BMI, correct percentage of body fat change through years. For example, men with 19% of body fat in his twenties, would have poor level of fitness, while same percentage for 50 years old man, would be excellent result. This problem cause Us to define one specific interval which we will use in further comprehensions. The easiest way to define a specific interval would be choosing the group with the largest amount of tested people.

```r
plot_norm  <- function(z, z_crit) {
    x_max  <- max(3, abs(z))
    x  <- seq(-x_max, x_max, 0.1)
    y  <- dnorm(x)
    plot(x, y, type="l")
    abline(v = z, col="blue", lty=2)
    for (zc in z_crit) {
        if (zc > 0) {
            x_c  <- c(zc, zc, x[x > zc], max(x))
            y_c  <- c(0, dnorm(zc), y[x > zc], 0) 
        } else {
            x_c  <- c(min(x), x[x < zc], zc, zc)
            y_c  <- c(0, y[x < zc], dnorm(zc), 0)             
        }
        polygon(x_c, y_c, border=NA, col="firebrick1")
    }
}
plot_norm(t, c(-z_kryt, z_kryt))
```

<h1>Test

To create our test we need to perform few calculations. First, as we know our sample group is active, hence they should confirm our hypotheses, that BMI indicator doesn't apply to physically active people. As we already know the average BMI of the test sample, we can try to predict with help of <i>Google</i> expected average Body Fat. So basically, we will calculate average BMI, estimate level of Body Fat which should match this BMI, and then we will test if our physically active sample will have lower level of Body Fat then estimated. This will confirm/deny our original hypotheses: BMI isn't accurate for physically active people.

<h3>Average BMI 

```r
mean(test_sample$bmi)
```

BMI at the level of 21.77 would result in average community in Body Fat at 23.5%. 

Our hipotez:
* $H_0: BF = 23.5$
* $H_1: BF <23.5$ 

```r
u0 <- 23.5
alfa <- 0.01
avg_bf <-mean(test_sample$bf)
sd_value <-sd(test_sample$bf)
n <-length(test_sample$bmi)
```

<h2>Test value:

```r
test_value = ((avg_bf - u0) / sd_value) * sqrt(n)
test_value
```

<h3> Critic zone:

```r
kryt_value = qt(1 - alfa, n-1)
kryt_value
```

```r
plot_crit  <- function(t, crit, df = 0) {
    dist  <- function(x, df) {
        if (df == 0)
            dnorm(x)
        else 
            dt(x, df)
    }
    x_max  <- max(3, abs(t))
    x  <- seq(-x_max, x_max, 0.1)
    y  <- dist(x, df)
    plot(x, y, type="l", main=paste(crit, " "))
    abline(v = t, col="blue", lty=2)
    for (c in crit) {
        if (c > 0) {
            x_c  <- c(c, c, x[x > c], max(x))
            y_c  <- c(0, dist(c, df), y[x > c], 0) 
        } else {
            x_c  <- c(min(x), x[x < c], c, c)
            y_c  <- c(0, y[x < c], dist(c, df), 0)             
        }
        polygon(x_c, y_c, border=NA, col="firebrick1")
    }
}
plot_t  <- function(t, alpha, df, type = "ne") {
    if (type == "ne") {
        t_crit  <- qt(alpha/2, df, lower.tail = FALSE)
        plot_crit(t, c(-t_crit, t_crit))
    } else {
        t_crit  <- qt(alpha, df, lower.tail = FALSE)
        plot_crit(t, ifelse(type == "gt", t_crit, -t_crit))
    }
}
```

```r
plot_t(test_value, 0.01, n-1, "lt")
```
